# Tutorial8: 使用Ascend C完成昇腾算子开发

本节以AddCustom算子为例，给出使用Ascend C完成算子开发的步骤.

分以下几步来实现：
1. 创建vscode应用，使用平台提供的msopgen镜像
2. 按照镜像中的文档完成算子项目创建、算子开发、编译和测试

建议使用1张910B NPU运行本教程。

## 1、拉取镜像并安装

1.1 创建作业

登录SCOW平台，进入AI集群，然后点击作业->创建作业->vscode



在创建vscode页面中，选择msopgen公共镜像，提交


待作业状态为Running，点击连接进入vscode应用


开启终端命令
