# Tutorial10: 使用mindformers框架完成Qwen1.5大模型的推理

本节以Qwen1.5-7B-Chat模型为例，给出其使用mindformers框架完成推理的步骤.

分以下几步来实现：
1. 创建vscode应用，使用平台提供的mindformersy镜像
2. 按照镜像中的文档完成大模型的推理

建议使用1张910B NPU运行本教程。

## 1、拉取镜像并安装

1.1 创建作业

登录SCOW平台，进入AI集群，然后点击作业->创建作业->vscode



在创建vscode页面中，选择mindformers公共镜像、选择code-server算法、选择Qwen1.5-7B-Chat作为基础模型，提交;
如果公共镜像中没有mindformers，则选择远程镜像并填写地址：app-store-images.pku.edu.cn/ascend/mindspore:openeuler-python3.10-cann8.0.rc3.beta1-mindspore2.4.0



待作业状态为Running，点击连接进入vscode应用


开启终端命令



按照run.md的提示进行模型推理
